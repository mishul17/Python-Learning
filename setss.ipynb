## Sets

Sets are built-in data type in python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, amd performing mathematical set operations like union, intersection, difference and symmetric difference.

```python
## create a set

my_set={1,2,3,4,5,6}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set([1,2,3,4,5,6])
print(my_empty_set)
```

```python
my_empty_set=set([1,2,3,6,4,5,6])
print(my_empty_set)
```

```python
## Basic set operations
## Adding and removing elements
my_set.add(7)
print(my_set)
```

```python
## Remove the elements from the set
my_set.remove(3)
print(my_set)
```

```python
my_set.discard(11)   ## discard will not show the error if the elements is not in the set
print(my_set)
```

```python
## pop method (whatever element is first inserted that element only will get removed.)
removed_element=my_set.pop()
print(removed_element)
print(my_set)
```

```python
## clear all the elements
my_set.clear()
print(my_set)
```

```python
## set membership test (chechking whether the given element belong to the particular set or not)
my_set={1,2,3,4,5}
print( 3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Operations
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## UNION (it will combine both the set)
union_set=set1.union(set2)
print(union_set)

## INTERSECTION (it will give the common element)
intersection_set=set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)  ## it will take the common element and update in the first set
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## Difference 
print(set1.difference(set2))
print(set2.difference(set1))
```

```python
## Symmetric Difference (common elements will be removed and remaining elements from both the set will be combined)
set1.symmetric_difference(set2)
```

```python
## Sets Method
set1={1,2,3}
set2={3,4,5}


## is subset
print(set1.issubset(set2))
print(set1.issuperset(set2))
```

```python
lst=[1,2,2,3,4,4,5]
set(lst)
```

```python
### Counting unique words in text
text="In this tutorial we are discussing about sets"
words=text.split()

## Convert list of words to set to get unique words
unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
