```python
## Arithmetic Operations
a=10
b=5
add_result=a+b #addition
sub_result=a-b #subtraction
mult_result=a*b #multiplication
div_result=a/b #divsion
floor_div_result=a//b #floor division it will remove the decimal give integer
modulus_result=a%b #modulus operation this will give the remainder
exponent_result=a**b #exponentiation
print(add_result)
print(sub_result)
print(mult_result)
print(div_result)
print(floor_div_result)
print(modulus_result)
print(exponent_result)




```

#### Logical Operator

```python
## and ( in the both have to be true)
X=True
Y=True
result=X and Y
print(result)
```

```python
# OR (if one of them is true then true)
X=False
Y=True
result=X or Y
print(result)
```
